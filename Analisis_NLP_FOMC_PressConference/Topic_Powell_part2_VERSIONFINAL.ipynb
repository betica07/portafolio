```python
media_list = [
    "CNN Business",
    "Politico",
    "The Washington Post",
    "Bloomberg Radio and Television",
    "Bloomberg News",
    "Associated Press",
    "Reuters",
    "Marketplace",
    "American Banker",
    "The Economist",
    "Yahoo Finance",
    "Wall Street Journal",
    "CNBC",
    "MarketWatch",
    "Financial Times",
    "New York Times",
    "CBS News",
    "Punchbowl News",
    "Axios",
    "Fox Business",
    "NPR",
    "ABC",
    "Jefferies Financial Group",
    "HuffPost"
]


# Function to find the media name in a given text
def find_media(text):
    for media in media_list:
        if media.lower() in text.lower():  # Case-insensitive matching
            return media
    return None  # Return None if no media name is found

# Apply the function to the 'content' column and create a new column for media names
interventions_df["media"] = interventions_df["content"].apply(find_media)

# Display the updated DataFrame
print(interventions_df)
```

```python
# Function to clean text in the "content" column
def clean_text(text):
    # Remove line breaks and replace them with spaces
    text = text.replace("\n", " ")

    # Generalized regex for human sounds and pauses
    human_sounds_regex = r"\b(?:[Mm]+[Hh]\s[Hh]|[Uu]+[Mm]|[Hh]+[Mm]+|[Aa]+[Hh]|[Oo]+[Hh]|[Ee]+[Rr]*|[Uu]+[Hh]+|[Mm]+|[Hh]+|[Uu]+|[Pp]+[Hh]+|[Er]+)\b"

    # Remove human sounds using the regex
    text = re.sub(human_sounds_regex, "", text)

    # Fix stuttering (e.g., "s-sample" -> "sample")
    text = re.sub(r"\b([a-zA-Z])-\1\b", r"\1", text)

    # Eliminar prefijos 'b' si existen y no están seguidos de ninguna letra
    text = re.sub(r'\b(b)\b', '', text)

    # Keep numbers and periods while removing other special characters
    text = re.sub(r"[^a-zA-Z0-9.\sáéíóúüñçÁÉÍÓÚÜÑÇ]", '', text)

    # Replace NaN values with an empty string
    if pd.isnull(text):
        text = ""

    # Eliminar palabras vacías (stopwords)
    stop_words = set(stopwords.words('english'))  # Usar la lista de stopwords en inglés
    text = ' '.join(word for word in text.split() if word not in stop_words)

    # Remove all hyphens and replace them with a single space
    text = re.sub(r"-", " ", text)

    # Remove all apostrophes
    text = re.sub(r"'", "", text)

    # Replace multiple spaces with a single space
    text = re.sub(r"\s+", " ", text).strip()

    return text

# Apply the cleaning function to each cell in the "content" column
interventions_df["content"] = interventions_df["content"].apply(clean_text)

interventions_df
```

```python
# Function to extract full journalist names and clean the questioning column
def process_row(row):
    # Regex to match the journalist's full uppercase name
    match = re.search(r"\b([A-Z]+(?:\s[A-Z]+)*)\b", row["questioning"])

    if match:
        journalist_name = match.group(1)  # Extract the full journalist name (first and last names)
        # Remove everything before and including the first occurrence of the journalist's name
        cleaned_questioning = re.sub(rf"^.*?{re.escape(journalist_name)}\b", journalist_name, row["questioning"], count=1).strip()
    else:
        journalist_name = None
        cleaned_questioning = row["questioning"]

    return pd.Series([journalist_name, cleaned_questioning])

# Apply the function to process each row
interventions_df[["journalist_name", "questioning"]] = interventions_df.apply(process_row, axis=1)

# Reorder columns to: media, journalist_name, questioning
interventions_df = interventions_df[["media", "journalist_name", "questioning"]]

interventions_df
```

```python
# Function to split questioning column into Q&A pairs
def split_questions_answers(row):
    journalist_name = row["journalist_name"]
    text = row["questioning"]

    # Initialize Q&A columns
    q_a_pairs = {}
    question_num = 1
    answer_num = 1

    # Regex pattern for splitting text
    pattern = rf"{re.escape(journalist_name)}|CHAIR POWELL"
    segments = re.split(pattern, text)

    # Iterate through segments and identify Q&A pairs
    i = 0
    while i < len(segments):
        segment = segments[i].strip()  # Remove leading and trailing spaces
        if not segment or segment == 0 or segment == ". .":  # Skip empty segments
            i += 1
            continue

        # If "CHAIR POWELL" is followed by journalist's name + "CHAIR POWELL", merge segments
        if (
            i + 1 < len(segments)
            and segments[i + 1].strip().startswith(journalist_name)
            and i + 2 < len(segments)
            and segments[i + 2].strip().startswith("CHAIR POWELL")
        ):
            # Merge current segment with the next two segments
            merged_segment = f"{segment} {segments[i + 1].strip()} {segments[i + 2].strip()}"
            q_a_pairs[f"A{answer_num}"] = merged_segment.strip()
            answer_num += 1
            i += 3  # Skip the merged segments
            continue

        # Assign segment to the correct Q/A column
        if i % 2 == 0:  # Even index = answer
            q_a_pairs[f"A{answer_num}"] = segment
            answer_num += 1
        else:  # Odd index = question
            q_a_pairs[f"Q{question_num}"] = segment
            question_num += 1

        i += 1

    return pd.Series(q_a_pairs)

# Apply the function to split questions and answers
split_df = interventions_df.apply(split_questions_answers, axis=1)

# Concatenate the original DataFrame with the split Q&A columns
result_df = pd.concat([interventions_df[["media", "journalist_name"]], split_df], axis=1)

# Reorder columns to: media, journalist_name, questioning
result_df = result_df[["media", "journalist_name", "Q1", "A1", "Q2", "A2", "Q3", "A3"]]
```

```python
# Function to calculate subjectivity based on sentiment score variance
# 0.0 --> objective
# 1.0 --> subjective

nlp = stanza.Pipeline('en', processors='tokenize,sentiment')
def calculate_subjectivity(sentiment_scores):
    if not sentiment_scores:
        return 0.0
    return sum(1 for score in sentiment_scores if score != 1) / len(sentiment_scores)  # Approximation

# Add subjectivity to the sentiment results before dropping
stanza_sentiment_df["subjectivity"] = stanza_sentiment_df["sentiment_scores"].apply(calculate_subjectivity)

# Drop the column outside the function
stanza_subjectivity_df = stanza_sentiment_df.drop(columns=["sentiment_scores"])

stanza_subjectivity_df
```

```python
from gensim.models import LdaModel
from gensim.corpora import Dictionary

nlp = stanza.Pipeline(lang='en', processors='tokenize,lemma')

# Define a custom stopword list
custom_stopwords = {'i', 'know', 'so', 'that', 'over', 'and', 'you', 'the', 'a', 'it', 'to', 'for', 'in', 'on', 'do', 'say', 'go', 'be', 'see', 'not', "'s", 'thank', 'thanks', 'chair'}

# Preprocess text using Stanza
def preprocess_text_stanza(text):
    if pd.isna(text) or not text.strip():  # Handle missing or empty text
        return []
    doc = nlp(text)
    tokens = []
    for sentence in doc.sentences:
        for word in sentence.words:
            if word.text.isalnum():  # Include words and numbers
                lemma = word.lemma.lower()
                if lemma not in custom_stopwords:  # Exclude stopwords
                    tokens.append(lemma)  # Use lemma for normalization
    return tokens

# Apply preprocessing to Q&A columns
corpus = []
for column in result_df.columns:
    if column.startswith("Q") or column.startswith("A"):  # Process only Q&A columns
        preprocessed_text = result_df[column].dropna().apply(preprocess_text_stanza)
        corpus.extend(preprocessed_text)

# Create Gensim dictionary and corpus
dictionary = Dictionary(corpus)
bow_corpus = [dictionary.doc2bow(doc) for doc in corpus]

# Train LDA model
lda_model = LdaModel(bow_corpus, num_topics=5, id2word=dictionary, passes=10)

# Print the topics
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx}: {topic}")
```

```python
nlp = stanza.Pipeline(lang='en', processors='tokenize,lemma')

# Define a custom stopword list
custom_stopwords = {'i', 'know', 'so', 'that', 'over', 'and', 'you', 'the', 'a', 'it', 'to', 'for', 'in', 'on', 'do', 'say', 'go', 'be', 'see', 'not', "'s", 'thank', 'thanks', 'chair', 'we'}

# Preprocess text using Stanza
def preprocess_text_stanza(text):
    if not text.strip():  # Handle empty text
        return []
    doc = nlp(text)
    tokens = []
    for sentence in doc.sentences:
        for word in sentence.words:
            if word.text.isalnum():  # Include words and numbers
                lemma = word.lemma.lower()
                if lemma not in custom_stopwords:  # Exclude stopwords
                    tokens.append(lemma)  # Use lemma for normalization
    return tokens

# Preprocess the cleaned_corpus
processed_corpus = [preprocess_text_stanza(sentence) for sentence in cleaned_corpus.split('.') if sentence.strip()]

# Create Gensim dictionary and Bag-of-Words corpus
dictionary = Dictionary(processed_corpus)
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_corpus]

# Train the LDA model
lda_model = LdaModel(bow_corpus, num_topics=5, id2word=dictionary, passes=10)

# Print the topics
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx}: {topic}")
```
