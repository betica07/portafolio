##Cambio a parametros categoricos

```python
alz["Gender"] = alz["Gender"].replace([0, 1],["Male", "Female"])
alz["Ethnicity"] = alz["Ethnicity"].replace([0, 1, 2, 3], ["Caucasian", "African American", "Asian", "Other"])
alz["EducationLevel"] = alz["EducationLevel"].replace([0, 1, 2, 3], ["None", "High School", "Bachelor", "Higher"])
alz["Smoking"] = alz["Smoking"].replace([0, 1], ["No", "Yes"])
alz["FamilyHistoryAlzheimers"] = alz["FamilyHistoryAlzheimers"].replace([0, 1], ["No", "Yes"] )
alz["CardiovascularDisease"] = alz["CardiovascularDisease"].replace([0, 1], ["No", "Yes"])
alz["Diabetes"] = alz["Diabetes"].replace([0, 1], ["No", "Yes"])
alz["Depression"] = alz["Depression"].replace([0, 1], ["No", "Yes"])
alz["HeadInjury"] = alz["HeadInjury"].replace([0, 1], ["No", "Yes"])
alz["Hypertension"] = alz["Hypertension"].replace([0, 1], ["No", "Yes"])
alz["MemoryComplaints"] = alz["MemoryComplaints"].replace([0, 1], ["No", "Yes"])
alz["BehavioralProblems"] = alz["BehavioralProblems"].replace([0, 1], ["No", "Yes"])
alz["Confusion"] = alz["Confusion"].replace([0, 1], ["No", "Yes"])
alz["Disorientation"] = alz["Disorientation"].replace([0, 1], ["No", "Yes"])
alz["PersonalityChanges"] = alz["PersonalityChanges"].replace([0, 1], ["No", "Yes"])
alz["DifficultyCompletingTasks"] = alz["DifficultyCompletingTasks"].replace([0, 1], ["No", "Yes"])
alz["Forgetfulness"] = alz["Forgetfulness"].replace([0, 1], ["No", "Yes"])
alz["Diagnosis"] = alz["Diagnosis"].replace([0, 1], ["No", "Yes"])

display(alz)
```

#Relacion incidencia del alzheimer y genero

```python
genre = alz[["Gender", "Diagnosis"]]
#display(genre)

gender_diagnosis_counts = genre.groupby('Gender')['Diagnosis'].value_counts()


alzheimers_counts = gender_diagnosis_counts.xs('Yes', level='Diagnosis')


plt.figure(figsize=(6, 6))
plt.pie(alzheimers_counts, labels=alzheimers_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Alzheimer\'s Incidence by Gender')
plt.show()
```

#Relacion incidencia y edad

```python
age = alz[["Age", "Diagnosis"]]

age_diagnosis_counts = age.groupby('Age')['Diagnosis'].value_counts()


alzheimers_by_age = age_diagnosis_counts.xs('Yes', level='Diagnosis')


ages = alzheimers_by_age.index
counts = alzheimers_by_age.values


plt.figure(figsize=(10, 8))
plt.bar(ages, counts, width=0.8, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of Alzheimer\'s Cases')
plt.title('Alzheimer\'s Incidence by Age')
plt.xticks(np.arange(min(ages), max(ages) + 1, 1.0))
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(10, 8))
colors = sns.color_palette("pastel")
num_colors = len(colors)

for i, v in enumerate(counts):
    plt.bar(ages[i], v, width=0.8, edgecolor='black', color=colors[i % num_colors]) #
    plt.text(ages[i], v + 0.2, str(v), ha='center', va='bottom', fontsize=10)

# ... (rest of the code)
plt.xlabel('Age')
plt.ylabel('Number of Alzheimer\'s Cases')
plt.title('Alzheimer\'s Incidence by Age')
plt.xticks(np.arange(min(ages), max(ages) + 1, 1.0))
plt.grid(True)
plt.show()
```

#Cantidad de pacientes con enfermedades subyacentes ademas del alzheimer

```python
under = alz[["CardiovascularDisease", "Hypertension", "Diabetes", "Depression", "HeadInjury", "Diagnosis"]]

import matplotlib.pyplot as plt
import numpy as np

diagnosed_patients = under[under['Diagnosis'] == 'Yes']

conditions = ['CardiovascularDisease', 'Hypertension', 'Diabetes', 'Depression', 'HeadInjury']

counts = [
    diagnosed_patients[condition].value_counts().get('Yes', 0)
    for condition in conditions
]

plt.figure(figsize=(8, 6))
plt.bar(conditions, counts, color='skyblue', edgecolor='black')
plt.xlabel('Underlying Conditions')
plt.ylabel('Number of Patients')
plt.title('Prevalence of Underlying Conditions in Alzheimer\'s Patients')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#Relacion incidencia con sintomas

```python
sintomas = alz[["MemoryComplaints","BehavioralProblems","Confusion", "Disorientation", "PersonalityChanges", "DifficultyCompletingTasks", "Forgetfulness", "Diagnosis"]]
#display(sintomas)

positivos = sintomas[sintomas['Diagnosis'] == 'Yes']



lifestyle = ["MemoryComplaints","BehavioralProblems","Confusion", "Disorientation", "PersonalityChanges", "DifficultyCompletingTasks", "Forgetfulness"]

conteo = [
    positivos[style].value_counts().get('Yes', 0)
    for style in lifestyle
]

plt.figure(figsize=(8, 6))
plt.bar(lifestyle, conteo, color='skyblue', edgecolor='black')
plt.xlabel('Sintoms')
plt.ylabel('Patient amount')
plt.title('Frecuency of sintoms in Alzheimer\'s Patients')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
